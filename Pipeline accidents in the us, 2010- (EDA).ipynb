```python
#построение графиков
#взято с 
#https://www.kaggle.com/thisherenow/oil-pipeline-spill-visualizations

group = df.groupby('Accident Year')

xPt = group.sum()['All Costs'].index.values
yAll_sum = group.sum()['All Costs'].values
yPro_sum = group.sum()['Property Damage Costs'].values
yCom_sum = group.sum()['Lost Commodity Costs'].values
yPri_sum = group.sum()['Public/Private Property Damage Costs'].values
yEme_sum = group.sum()['Emergency Response Costs'].values
yEnv_sum = group.sum()['Environmental Remediation Costs'].values

fracProp = group.mean()['Property Damage Costs'].values / group.mean()['All Costs'].values
fracComm = group.mean()['Lost Commodity Costs'].values / group.mean()['All Costs'].values
fracPriP = group.mean()['Public/Private Property Damage Costs'].values / group.mean()['All Costs'].values
fracEmer = group.mean()['Emergency Response Costs'].values / group.mean()['All Costs'].values
fracEnvi = group.mean()['Environmental Remediation Costs'].values / group.mean()['All Costs'].values
```

### Разделяем данные на тестовую и проверочную выборки

```python
from sklearn.model_selection import train_test_split
train,test,train_label,test_label=train_test_split(x,y,test_size=0.2,random_state=123)
```

### Нормализация Х_train и X_test

```python
#Импортируем модуль для стандартизации
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(train)

X_train_scaled = scaler.transform(train)
#print (X_train_scaled.mean(axis=0))
#print (X_train_scaled.std(axis=0))

X_test_scaled = scaler.transform(test)
#print (X_test_scaled.mean(axis=0))
#print (X_test_scaled.std(axis=0))
```

### Подбор параметров для случайного леса 

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
#Pipeline with preprocessing and modelPython
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV

pipeline = make_pipeline(preprocessing.StandardScaler(), 
                         RandomForestRegressor(n_estimators=100))
#Подбираем гипермапаметры

print (pipeline.get_params())
# ...
# 'randomforestregressor__criterion': 'mse',
# 'randomforestregressor__max_depth': None,
# 'randomforestregressor__max_features': 'auto',
# 'randomforestregressor__max_leaf_nodes': None,
# ...
#задаем словарь с гиперпараметрами
hyperparameters = { 'randomforestregressor__max_features' : ['auto', 'sqrt', 'log2'],
                  'randomforestregressor__max_depth': [None, 5, 3, 1]}
```

### Кросс-валидация и подгонка модели

```python
#кросс-валидация
clf = GridSearchCV(pipeline, hyperparameters, cv=10)
 
# Fit and tune model
clf.fit(train, train_label)
```

```python
#выводим лучшие параметры
print (clf.best_params_)
#проверяем подогнана ли модель
print (clf.refit)
```

### Предсказание и проверка метрик

```python
#предсказываем данные
y_pred = clf.predict(test)
```

```python
from sklearn import metrics
print ( 'R^2= ',metrics.r2_score(test_label, y_pred))
 
print ('MSE = ',metrics.mean_squared_error(test_label, y_pred))
```

```python
#собираем х и у для catboost
#зависимая переменная y
y = df_filled['All Costs']
#отберем иксы из датасета
x_ = df_filled[[
     'Unintentional Release (Barrels)',
     'Net Loss (Barrels)', 
    'Intentional Release (Barrels)', 
    'Liquid Recovery (Barrels)',
    'Property Damage Costs', 
    'Lost Commodity Costs',
    'Public/Private Property Damage Costs',
    'Emergency Response Costs',
    'Environmental Remediation Costs',
    'Other Costs'
]].astype('float64')
X = pd.concat([x_, df_filled[['Pipeline Location', 'Pipeline Type', 
                                    'Liquid Type', 'Cause Category',
                                   'Liquid Ignition','Liquid Explosion','Pipeline Shutdown']]], axis=1)

X.info()
```

```python
from sklearn.model_selection import train_test_split
train_data,test_data,train_label,test_label=train_test_split(X,y,test_size=0.1,random_state=123)
```

```python
from catboost import Pool, CatBoostRegressor
# определяем, какие данные будут категориальными

categorical_features_indices = np.where(X.dtypes != np.float)[0]

# создаем Pool
train_pool = Pool(train_data, 
                  train_label, 
                  cat_features=categorical_features_indices)
test_pool = Pool(test_data, 
                 cat_features=categorical_features_indices) 



# инициализируем CatBoostRegressor
model = CatBoostRegressor(n_estimators=100,
#                          num_trees=1000,
                          depth=7, 
                          learning_rate=0.1, 
                          loss_function='RMSE', 
                          custom_metric='R2',
                         early_stopping_rounds=5
)
# Fit model
model.fit(train_pool,eval_set=(test_data, test_label), plot=True)
# Get predictions
preds = model.predict(test_pool)
```

```python
print(model.get_best_score())
print(model.get_evals_result())
print(model.eval_metrics(train_pool,['R2']))
```
