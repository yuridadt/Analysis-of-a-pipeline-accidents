```python
# 
#  
#https://www.kaggle.com/thisherenow/oil-pipeline-spill-visualizations

group = df.groupby('Accident Year')

xPt = group.sum()['All Costs'].index.values
yAll_sum = group.sum()['All Costs'].values
yPro_sum = group.sum()['Property Damage Costs'].values
yCom_sum = group.sum()['Lost Commodity Costs'].values
yPri_sum = group.sum()['Public/Private Property Damage Costs'].values
yEme_sum = group.sum()['Emergency Response Costs'].values
yEnv_sum = group.sum()['Environmental Remediation Costs'].values

fracProp = group.mean()['Property Damage Costs'].values / group.mean()['All Costs'].values
fracComm = group.mean()['Lost Commodity Costs'].values / group.mean()['All Costs'].values
fracPriP = group.mean()['Public/Private Property Damage Costs'].values / group.mean()['All Costs'].values
fracEmer = group.mean()['Emergency Response Costs'].values / group.mean()['All Costs'].values
fracEnvi = group.mean()['Environmental Remediation Costs'].values / group.mean()['All Costs'].values
```

```python
f, a = plt.subplots(nrows=1, ncols=2, figsize=(13,5))

a[0].plot(xPt, yAll_sum, label="'All Costs", color='black')
a[0].plot(xPt, yEme_sum,label="Emergency Response Costs", color='orange')
a[0].plot(xPt, yEnv_sum,label="Environment", color='yellow')
a[0].plot(xPt, yPro_sum,label="Property Damage Costs", color='darkblue')
a[0].plot(xPt, yCom_sum,label="Lost Commodity Costs", color='darkred')
a[0].plot(xPt, yPri_sum,label="Public/Private Property Damage Costs.", color='darkgreen')
a[0].fill_between(xPt, yAll_sum, 10E0, facecolor='black', alpha=0.1)
a[0].fill_between(xPt, yPro_sum, 10E0, facecolor='darkblue', alpha=0.1)
a[0].fill_between(xPt, yCom_sum, 10E0, facecolor='darkred', alpha=0.1)
a[0].fill_between(xPt, yPri_sum, 10E0, facecolor='darkgreen', alpha=0.1)
a[0].fill_between(xPt, yEme_sum, 10E0, facecolor='orange', alpha=0.1)
a[0].fill_between(xPt, yEnv_sum, 10E0, facecolor='yellow', alpha=0.1)
a[0].get_xaxis().get_major_formatter().set_useOffset(False)
a[0].set_xlabel('Year', fontsize=16)
a[0].set_ylabel('Sum Cost ($Million)', fontsize=16)
a[0].legend()


a[1].plot(xPt, fracEmer,label="Emergency", color='orange')
a[1].plot(xPt, fracEnvi,label="Environment", color='yellow')
a[1].plot(xPt, fracProp,label="Property", color='darkblue')
a[1].plot(xPt, fracComm,label="Commodity", color='darkred')
a[1].plot(xPt, fracPriP,label="Public/Private Prop.", color='darkgreen')
a[1].fill_between(xPt, fracProp, 0, facecolor='darkblue', alpha=0.1)
a[1].fill_between(xPt, fracComm, 0, facecolor='darkred', alpha=0.1)
a[1].fill_between(xPt, fracPriP, 0, facecolor='darkgreen', alpha=0.1)
a[1].fill_between(xPt, fracEmer, 0, facecolor='orange', alpha=0.1)
a[1].fill_between(xPt, fracEnvi, 0, facecolor='yellow', alpha=0.1)
a[1].get_xaxis().get_major_formatter().set_useOffset(False)
a[1].set_xlabel('Year', fontsize=16)
a[1].set_ylabel('Fraction of the Total Cost', fontsize=16)
a[1].set_ylim([0,1])
tmp = a[1].legend()
```

```python
import powerlaw

dataCausMat=CausMat['All Costs'].values
dataCausMat=dataCausMat/10**3

fit = powerlaw.Fit(dataCausMat)
print(fit.power_law.alpha)
print(fit.power_law.xmin)
R, p = fit.distribution_compare('power_law', 'lognormal')
print(R,p)
```

```python
plt.figure(figsize=(15,8))

sns.set( context='poster', font_scale=0.7)

#  PDF
fig=plt.subplot(1, 2, 1)

figPDF = fit.plot_pdf(color='b', linewidth=2)
fit.power_law.plot_pdf(color='r', linestyle='--', ax=figPDF)
fit.lognormal.plot_pdf(color='g', linestyle='--', ax=figPDF)
#fit.plot_pdf(linear_bins=True,  color='r', ax=figPDF)
figPDF.set_title("PDF")
figPDF.set_ylabel("p(X)")
figPDF.set_xlabel(r"All Costs Frequency")

#  CCDF
plt.subplot(1, 2, 2)
figCCDF = fit.plot_ccdf(color='r', linewidth=2)
fit.power_law.plot_ccdf(color='r', linestyle='--', ax=figCCDF)
fit.lognormal.plot_ccdf(color='g', linestyle='--', ax=figCCDF)
####
figCCDF.set_title("CCDF")
figCCDF.set_ylabel(u" p(Xx)")
figCCDF.set_xlabel(r"All Costs Frequency")
```
