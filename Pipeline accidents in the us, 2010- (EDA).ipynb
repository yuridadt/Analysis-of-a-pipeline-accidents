```python
# 
#  
#https://www.kaggle.com/thisherenow/oil-pipeline-spill-visualizations

group = df.groupby('Accident Year')

xPt = group.sum()['All Costs'].index.values
yAll_sum = group.sum()['All Costs'].values
yPro_sum = group.sum()['Property Damage Costs'].values
yCom_sum = group.sum()['Lost Commodity Costs'].values
yPri_sum = group.sum()['Public/Private Property Damage Costs'].values
yEme_sum = group.sum()['Emergency Response Costs'].values
yEnv_sum = group.sum()['Environmental Remediation Costs'].values

fracProp = group.mean()['Property Damage Costs'].values / group.mean()['All Costs'].values
fracComm = group.mean()['Lost Commodity Costs'].values / group.mean()['All Costs'].values
fracPriP = group.mean()['Public/Private Property Damage Costs'].values / group.mean()['All Costs'].values
fracEmer = group.mean()['Emergency Response Costs'].values / group.mean()['All Costs'].values
fracEnvi = group.mean()['Environmental Remediation Costs'].values / group.mean()['All Costs'].values
```

```python
#     catboost
#  y
y = df_filled['All Costs']
#   
x_ = df_filled[[
     'Unintentional Release (Barrels)',
     'Net Loss (Barrels)', 
    'Intentional Release (Barrels)', 
    'Liquid Recovery (Barrels)',
    'Property Damage Costs', 
    'Lost Commodity Costs',
    'Public/Private Property Damage Costs',
    'Emergency Response Costs',
    'Environmental Remediation Costs',
    'Other Costs'
]].astype('float64')
X = pd.concat([x_, df_filled[['Pipeline Location', 'Pipeline Type', 
                                    'Liquid Type', 'Cause Category',
                                   'Liquid Ignition','Liquid Explosion','Pipeline Shutdown']]], axis=1)

X.info()
```

```python
from sklearn.model_selection import train_test_split
train_data,test_data,train_label,test_label=train_test_split(X,y,test_size=0.1,random_state=123)
```

```python
from catboost import Pool, CatBoostRegressor
# ,    

categorical_features_indices = np.where(X.dtypes != np.float)[0]

#  Pool
train_pool = Pool(train_data, 
                  train_label, 
                  cat_features=categorical_features_indices)
test_pool = Pool(test_data, 
                 cat_features=categorical_features_indices) 



#  CatBoostRegressor
model = CatBoostRegressor(n_estimators=100,
#                          num_trees=1000,
                          depth=7, 
                          learning_rate=0.1, 
                          loss_function='RMSE', 
                          custom_metric='R2',
                         early_stopping_rounds=5
)
# Fit model
model.fit(train_pool,eval_set=(test_data, test_label), plot=True)
# Get predictions
preds = model.predict(test_pool)
```

```python
print(model.get_best_score())
print(model.get_evals_result())
print(model.eval_metrics(train_pool,['R2']))
```
