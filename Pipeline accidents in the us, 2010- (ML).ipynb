### PCA

```python
from sklearn.decomposition import PCA
pca = PCA().fit(x)

#построим график выбора количества главных компонент
plt.figure(figsize=(10,7))
plt.plot(np.cumsum(pca.explained_variance_ratio_), color='k', lw=2)
plt.xlabel('Number of components')
plt.ylabel('Total explained variance')
plt.xlim(0, 63)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.axvline(5, c='b')
plt.axhline(0.9, c='r')
plt.show()
```

```python
pca = PCA(0.99).fit(x)
print('We need %d components to explain 99%% of variance' 
      % pca.n_components_)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=123)
print('Train size:', X_train.shape[0], 'Test size:', X_test.shape[0])
```

```python
pca = PCA(n_components=6)
pca.fit(X_train)

print('6 principal components explain %.2f%% of variance' %
      (100 * np.cumsum(pca.explained_variance_ratio_)[-1]))
plt.figure(figsize=(10,7))
plt.plot(np.cumsum(pca.explained_variance_ratio_), lw=2, color='k')
plt.xlabel('Number of components')
plt.ylabel('Total explained variance')
plt.xlim(0, 10)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.show()
```

#### Линейная регерссия в PCA

```python
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)

from sklearn.linear_model import LinearRegression
from sklearn import metrics

lin = LinearRegression(fit_intercept=True, normalize=True)
model = lin.fit(X_train_pca,y_train)
pred = model.predict(X_test_pca)

print('R^2: ', metrics.r2_score(y_test,pred))
print('MAE:', metrics.mean_absolute_error(y_test, pred))
print('MSE:', metrics.mean_squared_error(y_test, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pred)))
```

#### Деревья решений в PCA

```python
from sklearn import preprocessing
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV

pipeline = make_pipeline(preprocessing.StandardScaler(), 
                         RandomForestRegressor(n_estimators=100))

hyperparameters = { 'randomforestregressor__max_features' : ['auto', 'sqrt', 'log2'],
                  'randomforestregressor__max_depth': [None, 5, 3, 1]}
```

```python
#кросс-валидация
clf = GridSearchCV(pipeline, hyperparameters, cv=10)
 
# Fit and tune model
clf.fit(X_train_pca, y_train)

#выводим лучшие параметры
print (clf.best_params_)
#проверяем подогнана ли модель
print (clf.refit)
```

```python
#предсказываем данные
y_pred = clf.predict(X_test_pca)

print('R^2: ', metrics.r2_score(y_test,y_pred))
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
from catboost import Pool, CatBoostRegressor
# определяем, какие данные будут категориальными

categorical_features_indices = np.where(X.dtypes != np.float)[0]

# создаем Pool
train_pool = Pool(train_data, 
                  train_label, 
                  cat_features=categorical_features_indices)
test_pool = Pool(test_data, 
                 cat_features=categorical_features_indices) 



# инициализируем CatBoostRegressor
model = CatBoostRegressor(n_estimators=100,
#                          num_trees=1000,
                          depth=7, 
                          learning_rate=0.1, 
                          loss_function='RMSE', 
                          custom_metric='R2',
                         early_stopping_rounds=5
)
# Fit model
model.fit(train_pool,eval_set=(test_data, test_label), plot=True)
# Get predictions
preds = model.predict(test_pool)
```

```python
print(model.get_best_score())
print(model.get_evals_result())
print(model.eval_metrics(train_pool,['R2']))
```
